# CONSTANT FUNCTIONS

## f(x) = 0

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram

def Deutsch_Josza_algorithm(oracle_gate):
    qc = QuantumCircuit(4, 3)
    qc.x(3)
    qc.h(range(4))
    qc.append(oracle_gate, range(4))
    qc.h(range(3))
    qc.measure([0, 1, 2], [0, 1, 2])
    qc.draw('mpl')
    return qc

def oracle_constant():
    qc = QuantumCircuit(4)
    return qc

oracle_gate_constant = oracle_constant().to_gate()
oracle_gate_constant.name = "Constant function"
dj_circuit_constant = Deutsch_Josza_algorithm(oracle_gate_constant)
aer_sim = Aer.get_backend('aer_simulator')
transpiled_dj_circuit_constant = transpile(dj_circuit_constant, aer_sim)
qobj_constant = assemble(transpiled_dj_circuit_constant)
results_constant = aer_sim.run(qobj_constant).result()
answer_constant = results_constant.get_counts()
plot_histogram(answer_constant)
```
